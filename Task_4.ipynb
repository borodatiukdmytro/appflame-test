```python
import pandas as pd
data = pd.read_excel('Data_4.xlsx')
data.head(20)
```

Виведення всіх унікальних значень дати, для того щоб прибрати нерелевантні

```python
unique_values = data['TD'].unique()
print(unique_values)
```

чистка даних по даті

```python
df = data[data['TD'] != '(null)']
df
```

виведення всіх варіантів сум прибутку, для того щоб прибрати нерелевантні

```python
unique_values = df['revenue_usd'].unique()
print(unique_values)
```

чистка даних по доходу

```python
df['revenue_usd'] = df['revenue_usd'].replace('(null)', 0)
df
```

виведення кількості унікальних клієнтів

```python
print(df['customer_id'].nunique())
```

створення таблиці когортного аналізу

```python
df['TD'] = pd.to_datetime(df['TD'])

# Визначення когорти за місяцем
df['CohortMonth'] = df.groupby('customer_id')['TD'].transform('min').dt.to_period('M')

# Іизначення к-ті місяців з першої транзакції
df['MonthsSinceCohort'] = (df['TD'].dt.to_period('M') - df['CohortMonth']).apply(lambda x: x.n) + 1

cohort_pivot = df.pivot_table(
    index='CohortMonth',  # Когорта - місяць 1 підписки
    columns='MonthsSinceCohort',  
    values='revenue_usd',  # Сумма доходу
    aggfunc='sum'
)

cohort_pivot
```

```python
# Calculate cumulative revenue
cohort_pivot_cumulative = cohort_pivot.cumsum(axis=1)
cohort_pivot_cumulative
```

```python
# Calculate number of unique customers in each cohort
cohort_sizes = df.groupby('CohortMonth')['customer_id'].nunique()
cohort_sizes
```

```python
# Normalize revenue by cohort size
cohort_pivot_normalized = cohort_pivot_cumulative.div(cohort_sizes, axis=0)
cohort_pivot_normalized
```

```python
LTV = cohort_pivot_normalized.mean()
cohort_pivot_normalized.loc['LTV'] = LTV
cohort_pivot_normalized
```

```python
LTV_6th_month = LTV[6]
print(f"LTV for the 6th month: {LTV_6th_month}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

time_periods = list(range(1, len(LTV) + 1))

sns.set(style='whitegrid')

plt.figure(figsize=(12, 8))
plt.plot(time_periods, LTV, marker='o', linestyle='-', color='teal', markersize=8, linewidth=2, label='LTV')

plt.title('Крива LTV за весь період', fontsize=16, fontweight='bold')
plt.xlabel('Time Period', fontsize=14)

plt.legend()

for i, value in enumerate(LTV):
    plt.text(time_periods[i], value, f'{value:.2f}', fontsize=12, ha='right', va='bottom')

mean_ltv = sum(LTV) / len(LTV)
plt.axhline(y=mean_ltv, color='r', linestyle='--', label=f'Mean LTV ({mean_ltv:.2f})')

plt.text(x=max(time_periods)*0.9, y=mean_ltv, s=f'Mean LTV ({mean_ltv:.2f})', color='r', fontsize=12, va='bottom', ha='right')

plt.grid(True)
plt.xticks(ticks=time_periods)

plt.tight_layout()
plt.show()
```
